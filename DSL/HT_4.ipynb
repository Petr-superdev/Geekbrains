#### Задание 2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080', 'family': 'Calibri'}
label_dict = {'fontsize': 16, 'color': '#808080', 'family': 'Calibri'}

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.title('График f(t)', fontdict=title_dict)

plt.xlabel('Значения t', fontdict=label_dict, loc='right')
plt.ylabel('Значения f', fontdict=label_dict, loc='top')

plt.plot(t, f, color='green')
plt.show()
```

#### Задание 3

```python
x = np.linspace(-3, 3, 51)

y1 = x ** 2
y2 = 2 * x + 0.5
y3 = 3 * x - 1.5
y4 = np.sin(x)
```

```python
title_dict = {'fontsize': 10, 'fontweight': 'bold', 'color': '#808080', 'family': 'Calibri'}

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1', fontdict=title_dict)
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2', fontdict=title_dict)

ax3.plot(x, y3)
ax3.set_title('График y3', fontdict=title_dict)

ax4.plot(x, y4)
ax4.set_title('График y4', fontdict=title_dict)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
